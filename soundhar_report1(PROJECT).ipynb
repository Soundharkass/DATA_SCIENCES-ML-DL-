# USING **HEATMAP** PLOTS

```python
DT = dt.drop(columns=["Unnamed: 0"])
print(DT)

#columns=["Unnamed: 0"]: Specifies the column you want to remove.
#Unnamed: 0 is typically an accidental index column created when saving a DataFrame to a CSV file without setting index=False.
#dt.drop(...): Deletes that specific column from your data table and creates a clean copy named DT.
#print(DT): Displays the updated, cleaner table in your console.
```

```python
corr = DT.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()

#DT.corr(): Calculates the correlation coefficients between all numerical columns, 
#showing how strongly pairs of variables change together (ranging from -1 to +1).
#sns.heatmap(...): Visualizes these correlation numbers as a color-coded grid, 
#making strong relationships easy to spot instantly.
#annot=True: Overlays the actual mathematical correlation values inside each square so you can read the exact numbers.
#cmap="coolwarm": Sets the color theme where warm colors (red) represent strong positive correlations and,
#cool colors (blue) represent negative correlations.
```
