# USING **HEATMAP** PLOTS

```python
DT = dt.drop(columns=["Unnamed: 0"])
print(DT)
```

```python
corr = DT.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()
```
